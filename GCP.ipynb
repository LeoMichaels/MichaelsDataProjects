```python
df['Usage Start Date'] = pd.to_datetime(df['Usage Start Date'], errors='coerce')
df['Usage End Date'] = pd.to_datetime(df['Usage End Date'], errors='coerce')
```

```python
df['Duration (hours)'] = (df['Usage End Date'] - df['Usage Start Date']).dt.total_seconds() / 3600
```

```python
df.head(5)
```

```python
df['Duration (hours)'] = df['Duration (hours)'].replace(0, 1)
df['Rounded Cost ($)'] = df['Rounded Cost ($)'].replace(0, 1)
```

```python
df['Network Total (GB)'] = (df['Network Inbound Data (Bytes)'] + df['Network Outbound Data (Bytes)']) / (1024 ** 3)
```

```python
df.head(5)
```

```python
# Optional: Add weight to network usage
df['Usage Factor'] = df['CPU Utilization (%)'] + df['Memory Utilization (%)'] + (df['Network Total (GB)'] * 5)

# Compute time_value_utility
df['time_value_utility'] = df['Usage Factor'] / (df['Rounded Cost ($)'] * df['Duration (hours)'])
```

```python
df.head(5)
```

```python
df.to_excel(r"C:\Users\leomi\OneDrive\Desktop\Data\GPL Cloud Data\GPL_TVU_RunningCopy.xlsx")
```

```python
closeup = pd.DataFrame()

closeup = df.groupby(["Service Name"]).agg(
    total_rounded_cost = ("Rounded Cost ($)", "sum"),
    median_time_value = ("time_value_utility", "median")).reset_index()
```

```python
closeup.head(2)
```

```python
closeup["Value_Class"] = pd.qcut(closeup["median_time_value"], q=3, labels=["Low", "Medium", "High"])
```
